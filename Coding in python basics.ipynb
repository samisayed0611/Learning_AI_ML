# Example 1

Print the prime numbers from 1 to 100.

```python
# Example 1
for i in range(101):
  prime=True
  j=2
  while prime and j<i:
    if (i % (j))==0:
      prime=False
    j+=1
  if prime:
    print(i)
```

$$ y = \frac{y_{max} - y_{min}}{x_{max}-x_{min}}(x - x_{min}) + y_{min} $$

where $x$ is raw value and $y$ is the normalised value. The $y_{min}$ and $y_{max}$ determine the output range. The $x_{max}$ and $x_{min}$ denote the input range.

For this task $y_{max}=1$, $y_{min}=0$. The $x_{max}$ and $x_{min}$ must be obtained from the dataset, for each column separately.

# Task 9

The sigmoid function is frequently used activation of a neuron. It is defined by

$$y=\sigma(x)=\frac{1}{1+e^{-x}}$$

Draw a graf of this function for $x\in <-8,8>$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values from -8 to 8
x = np.linspace(-8, 8, 400)

# Calculate y values using the sigmoid function
y = sigmoid(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sigmoid Function', color='b')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()

```

# Task 10

The RELU function is frequently used activation function of neurons in deep neural networks. It is defined by

$$y=\mbox{relu}(x)=\mbox{max}(0, x)$$

Draw a graf of this function for $x\in <-10,10>$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the ReLU function
def relu(x):
    return np.maximum(0, x)

# Generate x values from -10 to 10
x = np.linspace(-10, 10, 400)

# Calculate y values using the ReLU function
y = relu(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='ReLU Function', color='b')
plt.title('ReLU Function')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.grid(True)
plt.legend()
# plt.axhline(y=0, color='k', linewidth=0.5)
# plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```

# Task 11

The Gaussian function is frequently used as an activation function of RBF neurons. It is defined by the equation

$$y=g(x)=e^{\frac{-x^2}{2\sigma^2}}$$

Draw a graf of this function for $x\in <-10,10>$ and $\sigma=\{0.2,0.5,1,2,5\}$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Gaussian function
def gaussian(x, sigma):
    return np.exp(-x**2 / (2 * sigma**2))

# Generate x values from -10 to 10
x = np.linspace(-10, 10, 400)

# Different values of sigma
sigmas = [0.2, 0.5, 1, 2, 5]

# Create a plot for each sigma
plt.figure(figsize=(8, 4))
for sigma in sigmas:
    y = gaussian(x, sigma)
    plt.plot(x, y, label=f'Sigma = {sigma}')

plt.title('Gaussian Function for Different Sigma Values')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.show()
```
